```python
df_t = df.copy()
df_t_1 = df.copy()
```

```python
df_t['회계년도_t-1'] = df_t['회계년도'] -1
```

```python
df_t_1.drop(['차기부도여부'], axis=1, inplace=True)
```

```python
for col in value_cc:
  df_t_1.loc[df_t_1[col]==0, col] = 0.000001
```

```python
merged_df = pd.merge(df_t, df_t_1, how='inner', left_on = ['산업코드', '소속코드','상장일','감사의견코드','산업대분류','회사명', '거래소코드','회계년도_t-1'], right_on = ['산업코드', '소속코드','상장일','감사의견코드','산업대분류','회사명', '거래소코드','회계년도'], suffixes = ('_t', '_t-1') )
```

```python
col_list = merged_df.columns

matching = [s for s in col_list if "회계년도" in s] 
print(matching)
```

```python
merged_df.drop(['회계년도_t-1', '회계년도_t-1'],axis=1, inplace=True)
```

```python
merged_df
```

```python
diff_df = pd.DataFrame()

diff_df['회사명'] = merged_df['회사명']
diff_df['거래소코드'] = merged_df['거래소코드']
diff_df['회계년도'] = merged_df['회계년도_t']
diff_df['산업대분류'] = merged_df['산업대분류']
diff_df['차기부도여부'] = merged_df['차기부도여부']

# 값을 가지는 컬럼 -> 성장률로 변환
for col in value_cc:
  diff_df['growth_rate_{}'.format(col)] = (merged_df.loc[:,col+'_t']- merged_df.loc[:,col+'_t-1']) / merged_df.loc[:,col+'_t-1']

# 비율을 가지는 컬럼
for col in value_bc:
  diff_df['growth_rate_{}'.format(col)] = merged_df.loc[:,col+'_t']- merged_df.loc[:,col+'_t-1']
```
